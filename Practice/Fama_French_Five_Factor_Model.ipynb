```python
import pandas as pd
import yfinance as yf
from pandas_datareader import data as pdr
```

# 1. Use Pre-calculated Factor

## Step 1: Data Collection

- [Fama-French Data (Ken French’s Data Library) — pandas-datareader 0.10.0 documentation](https://pandas-datareader.readthedocs.io/en/latest/readers/famafrench.html)
- [Fama-French Data (Ken French’s Data Library) — pandas-datareader v0.10.0 documentation](https://pydata.github.io/pandas-datareader/devel/readers/famafrench.html)

```python
# Load Fama-French 5 factors data
# FutureWarning: The argument 'date_parser' is deprecated and will be removed in a future version. Please use 'date_format' instead, or read your data in as 'object' dtype and then call 'to_datetime'.
ff_factors = pdr.get_data_famafrench('F-F_Research_Data_5_Factors_2x3', start='2010', end='122020')[0] / 100
ff_factors
```

```python
import calendar
ff_factors.index = ff_factors.index.to_timestamp().map(lambda x: x.replace(day=calendar.monthrange(x.year, x.month)[1]))
ff_factors
```

```python
# Load stock data (example for a set of tickers)
tickers = ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'NVDA']
stock_data = yf.download(tickers, start='2010-01-01', end='2020-12-31')['Adj Close']
stock_data
```

```python
# https://pandas.pydata.org/pandas-docs/version/1.4.0/user_guide/timeseries.html
# https://stackoverflow.com/questions/62917882/convert-datetime64ns-utc-pandas-column-to-datetime
stock_data.index = stock_data.index.tz_localize(None)
stock_data
```

```python
# Calculate daily returns
stock_returns = stock_data.pct_change().dropna()
stock_returns
```

```python
# Align with factor data (convert monthly factor data to daily for simplicity)
# FutureWarning: Resampling with a PeriodIndex is deprecated. Cast index to DatetimeIndex before resampling instead. (fixed)
factor_data = ff_factors.resample('D').ffill()
factor_data
```

## Step 2: Calculate Excess Returns

```python
# Assume 1-month Treasury bill rate as risk-free rate
risk_free_rate = factor_data['RF']
risk_free_rate
```

```python
excess_stock_returns = stock_returns.subtract(risk_free_rate, axis=0).dropna()
excess_stock_returns
```

## Step 3: Estimate Factor Exposures

```python
%pip install statsmodels
```

```python
factor_data = factor_data.loc[excess_stock_returns.index]
factor_data
```

```python
import statsmodels.api as sm

def regress_factors(excess_returns: pd.Series, factors: pd.DataFrame):
    X = sm.add_constant(factors[['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA']])
    model = sm.OLS(excess_returns, X)
    results = model.fit()
    return results

factor_loadings = {}
for ticker in tickers:
    results = regress_factors(excess_stock_returns[ticker], factor_data)
    print(results.summary())
    factor_loadings[ticker] = results.params

factor_loadings_df = pd.DataFrame(factor_loadings).T
factor_loadings_df
```

## Step 4: Calculate Stock Weights (Testing)

```python
# Example: Calculate scores based on factor loadings (e.g., prioritize profitable and conservative stocks)
factor_weights = {'Mkt-RF': 0.2, 'SMB': 0.1, 'HML': 0.1, 'RMW': 0.3, 'CMA': 0.3}

# Calculate stock scores based on weighted exposure to factors
stock_scores = factor_loadings_df.mul(pd.Series(factor_weights), axis=1).sum(axis=1)

# Normalize the scores to get weights
stock_weights = stock_scores / stock_scores.sum()

stock_weights
```

## Step 5: Backtest and Create the DataFrame with Weights

```python
future_ff_factors = pdr.get_data_famafrench('F-F_Research_Data_5_Factors_2x3', start='2021', end='2024')[0] / 100
future_ff_factors.index = future_ff_factors.index.to_timestamp().map(lambda x: x.replace(day=calendar.monthrange(x.year, x.month)[1]))
future_ff_factors
```

```python
# Align future factors with the same structure as factor loadings
future_factors_subset = future_ff_factors.drop('RF', axis=1)
future_factors_subset['const'] = 1.0
future_factors_subset
```

```python
# Calculate stock scores for each future date
# Multiply factor loadings with future factor values
stock_scores_future = future_factors_subset.dot(factor_loadings_df.T)
stock_scores_future
```

```python
# Normalize the scores to create stock weights
stock_weights_future = stock_scores_future.div(stock_scores_future.sum(axis=1), axis=0)
stock_weights_future
```

## Step 6: Portfolio Metrics Analysis

```python
# Load future stock data
future_stock_data = yf.download(tickers, start='2021-01-01', end='2024-01-31')['Adj Close']
future_stock_data.index = future_stock_data.index.tz_localize(None)

# Calculate daily returns
future_stock_returns = future_stock_data.pct_change().dropna()
future_stock_returns
```

```python
future_weighted_returns = (stock_weights_future * future_stock_returns).dropna()
future_weighted_returns
```

```python
# Portfolio returns
portfolio_returns = future_weighted_returns.sum(axis=1)
portfolio_returns
```

```python
from dateutil.relativedelta import relativedelta
# https://stackoverflow.com/questions/24284342/insert-a-row-to-pandas-dataframe
portfolio_returns.loc[portfolio_returns.index[0] - relativedelta(months=1)] = 0.0
portfolio_returns.sort_index(inplace=True)
portfolio_returns
```

```python
# Calculate metrics like cumulative return, volatility, Sharpe ratio
cumulative_returns = (1 + portfolio_returns).cumprod()
volatility = portfolio_returns.std() * (252 ** 0.5)
sharpe_ratio = portfolio_returns.mean() / portfolio_returns.std() * (252 ** 0.5)

print(f"Cumulative Return: {cumulative_returns.iloc[-1] - 1:.2f}")
print(f"Annualized Volatility: {volatility:.2f}")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
```

```python
# Download S&P 500 index data for the market return (you can also use a broader index)
future_market_data = yf.download('^GSPC', start='2021-01-01', end='2024-01-31')['Adj Close']
future_market_data.index = future_market_data.index.tz_localize(None)

# Calculate daily returns
future_market_returns = future_market_data.pct_change().dropna()
future_market_returns
```

```python
future_market_returns = future_market_returns.loc[future_weighted_returns.index]
future_market_returns
```

```python
future_market_returns.loc[future_market_returns.index[0] - relativedelta(months=1)] = 0.0
future_market_returns.sort_index(inplace=True)
market_cumulative_returns = (1 + future_market_returns).cumprod()
market_volatility = future_market_returns.std() * (252 ** 0.5)
market_sharpe_ratio = future_market_returns.mean() / future_market_returns.std() * (252 ** 0.5)

print(f"Market Cumulative Return: {market_cumulative_returns.iloc[-1] - 1:.2f}")
print(f"Market Annualized Volatility: {market_volatility:.2f}")
print(f"Market Sharpe Ratio: {market_sharpe_ratio:.2f}")
```

```python
import plotly.graph_objects as go

# Create a line plot of the NAV curve
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=cumulative_returns.index,  # Dates on the x-axis
    y=cumulative_returns,        # NAV on the y-axis
    mode='lines',
    name='Net Asset Value',
    # line=dict(color='blue')
))

fig.add_trace(go.Scatter(
    x=market_cumulative_returns.index,  # Dates on the x-axis
    y=market_cumulative_returns,        # NAV on the y-axis
    mode='lines',
    name='Baseline (Market (S&P 500))',
    # line=dict(color='blue')
))

# Add titles and labels
fig.update_layout(
    title="Net Asset Value (NAV) Curve",
    xaxis_title="Date",
    yaxis_title="NAV",
    template='plotly_dark'
)

# Show the plot
fig.show()
```
